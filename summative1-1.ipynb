5. Convert `Date` to a pandas timestamp. 
   > tip: use `pd.to_datetime()`

````python
# your code here
attendance['Date'] = pd.to_datetime(attendance['Date'])

````

6. Filter the DataFrame on **one specific module** (your choice) and plot its **attendance rate over time**
   (x = Date, y = the module's average attendance).



````python
# Filtering data for Arabic module only
Arabic_attendance = attendance[attendance['Module Name'] == 'Arabic']


#Calculating rate over time where x is date and y is module average attendance
attendance_rate = (Arabic_attendance.groupby('Date')['Attended'].mean())
    
#plot attendance rate over time
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(attendance_rate.index, attendance_rate.values, marker='o', linestyle='-', linewidth=2)
plt.title('Attendance Rate for Arabic Module Over Time')
plt.xlabel('Date')
plt.ylabel('Attendance Rate')
plt.show()

````

7. **[For L6 students only]** Add a column, titled "Student Overall Attendance", with the average attendance for the student. Add a column, titled "Standardised Student Overall Attendance", with the z-scores of "Student Overall Attendance".

````python
#Add a column for student overall attendance
attendance['Student Overall Attendance'] = attendance.groupby('Person code')['Attended'].transform('mean')
#Add a column for the standarised code with z-score of students
attendance['Standardised Student Attendance '] = (attendance['Student Overall Attendance'] - attendance['Student Overall Attendance'].mean()) / attendance['Student Overall Attendance'].std()

attendance.head()
````
